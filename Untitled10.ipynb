````python
import re
import os
import json
output_folder = 'nextjs_project'
drive_output_folder = '/content/drive/MyDrive/nextjs_project'
project_path = '/content/drive/MyDrive/angular-mobile-app'

# Initialize metadata
metadata = {
    "project_summary": "",
    "files": [],
    "components": [],
    "services": [],
    "routing": {"module": "", "routes": []},
    "modules": [],
    "configs": {},
    "assets": [],
    "stats": {"file_count": 0, "component_count": 0, "service_count": 0, "size_mb": 0}
}

# Helper: Categorize file
def categorize_file(path):
    if path.endswith('.component.ts'):
        return 'component'
    elif path.endswith('.service.ts'):
        return 'service'
    elif path.endswith('.module.ts'):
        return 'module'
    elif path.endswith(('.json', '.js', '.ts')) and ('environment' in path or 'config' in path.lower()):
        return 'config'
    elif path.endswith(('.png', '.jpg', '.svg', '.css', '.scss')):
        return 'asset'
    elif path.endswith(('.ts', '.js')):
        return 'utility'
    return 'other'

# Helper: Summarize content
def summarize_content(path, content, file_type):
    filename = os.path.basename(path)
    summary = ""

    if file_type == 'component':
        bindings = re.findall(r"\*ng\w+=\"[^\"]+\"|\{\{[^}]+}}|\(\w+\)=\"[^\"]+\"", content) or []
        methods = re.findall(r"\b(\w+)\([^)]*\)\s*{", content) or []
        selector = re.search(r"selector:\s*['\"]([^'\"]+)['\"]", content)
        template = re.search(r"templateUrl:\s*['\"]([^'\"]+)['\"]|template:\s*`", content)
        summary = f"Angular component '{filename}'. "
        if selector:
            summary += f"Selector: {selector.group(1)}. "
        if template:
            summary += "Renders UI with template. "
        if bindings:
            summary += f"Bindings: {', '.join(bindings[:2])}. "
        if methods:
            summary += f"Methods: {', '.join(methods[:2])}. "

    elif file_type == 'service':
        methods = re.findall(r"\b(\w+)\([^)]*\)\s*{", content) or []
        http = re.findall(r"http\.(get|post|put|delete)\b", content) or []
        observables = bool(re.search(r"\.pipe\(|Observable<", content))
        summary = f"Angular service '{filename}'. "
        if methods:
            summary += f"Methods: {', '.join(methods[:2])}. "
        if http:
            summary += f"HTTP: {', '.join(http[:2])}. "
        if observables:
            summary += "Uses observables. "

    elif file_type == 'module':
        declarations = re.findall(r"declarations:\s*\[\s*([^\]]+)\]", content) or []
        imports = re.findall(r"imports:\s*\[\s*([^\]]+)\]", content) or []
        providers = re.findall(r"providers:\s*\[\s*([^\]]+)\]", content) or []
        summary = f"Angular module '{filename}'. "
        if declarations:
            summary += f"Declares: {declarations[0][:50]}... "
        if imports:
            summary += f"Imports: {imports[0][:50]}... "
        if providers:
            summary += f"Provides: {providers[0][:50]}... "

    elif file_type == 'config':
        if 'package.json' in path:
            try:
                pkg = json.loads(content)
                deps = list(pkg.get('dependencies', {}).keys())[:2]
                summary = f"Config '{filename}': Defines dependencies ({', '.join(deps or ['none'])})."
            except:
                summary = f"Config '{filename}': Project metadata."
        elif 'angular.json' in path:
            summary = f"Config '{filename}': Angular build settings."
        elif 'environment' in path:
            summary = f"Config '{filename}': Environment variables."
        else:
            summary = f"Config '{filename}': Project settings."

    elif file_type == 'asset':
        ext = filename.split('.')[-1]
        role = "image" if ext in ('png', 'jpg', 'svg') else "stylesheet"
        summary = f"Asset '{filename}': {role.capitalize()} for app."

    elif file_type == 'utility':
        classes = re.findall(r"class (\w+)", content) or []
        functions = re.findall(r"\b(\w+)\([^)]*\)\s*{", content) or []
        summary = f"Utility '{filename}'. "
        if classes:
            summary += f"Classes: {', '.join(classes[:2])}. "
        if functions:
            summary += f"Functions: {', '.join(functions[:2])}. "

    else:
        if 'package-lock.json' in path:
            summary = f"'{filename}': Locks dependency versions."
        elif '.gitignore' in path:
            summary = f"'{filename}': Specifies ignored files."
        elif path.endswith('.html'):
            elements = re.findall(r"<(\w+)", content) or []
            summary = f"HTML '{filename}': Template with elements ({', '.join(elements[:2] or ['none'])})."
        elif path.endswith('.spec.ts'):
            summary = f"Test '{filename}': Unit tests."
        else:
            summary = f"'{filename}': Miscellaneous file in {os.path.dirname(path)}."

    return summary.strip() or f"'{filename}': Purpose unclear."

# Helper: Parse component
def parse_component(path, content, summary):
    component = {
        "name": "",
        "path": path,
        "selector": "",
        "template": "",
        "inputs": [],
        "outputs": [],
        "methods": [],
        "dependencies": [],
        "summary": summary
    }
    name_match = re.search(r'class (\w+)', content)
    if name_match:
        component["name"] = name_match.group(1)
    selector_match = re.search(r"selector:\s*['\"]([^'\"]+)['\"]", content)
    if selector_match:
        component["selector"] = selector_match.group(1)
    template_match = re.search(r"templateUrl:\s*['\"]([^'\"]+)['\"]", content)
    if template_match:
        component["template"] = template_match.group(1)
    else:
        template_inline = re.search(r"template:\s*`([^`]+)`", content)
        if template_inline:
            component["template"] = "inline"
    component["inputs"] = re.findall(r"@Input\(\)\s*(\w+)", content) or []
    component["outputs"] = re.findall(r"@Output\(\)\s*(\w+)", content) or []
    component["methods"] = re.findall(r"\b(\w+)\([^)]*\)\s*{", content) or []
    component["dependencies"] = re.findall(r"constructor\([^)]*(\w+Service)", content) or []
    return component

# Helper: Parse service
def parse_service(path, content, summary):
    service = {
        "name": "",
        "path": path,
        "methods": [],
        "http": False,
        "dependencies": [],
        "summary": summary
    }
    name_match = re.search(r'class (\w+)', content)
    if name_match:
        service["name"] = name_match.group(1)
    service["methods"] = re.findall(r"\b(\w+)\([^)]*\)\s*{", content) or []
    service["http"] = bool(re.search(r"http\.(get|post|put|delete)", content))
    service["dependencies"] = re.findall(r"constructor\([^)]*(\w+)", content) or []
    return service

# Helper: Parse module
def parse_module(path, content, summary):
    module = {
        "name": "",
        "path": path,
        "imports": [],
        "declarations": [],
        "providers": [],
        "summary": summary
    }
    name_match = re.search(r'class (\w+)', content)
    if name_match:
        module["name"] = name_match.group(1)
    module["imports"] = re.findall(r"imports:\s*\[\s*([^\]]+)\]", content) or []
    module["declarations"] = re.findall(r"declarations:\s*\[\s*([^\]]+)\]", content) or []
    module["providers"] = re.findall(r"providers:\s*\[\s*([^\]]+)\]", content) or []
    return module

# Project summary
def generate_project_summary():
    summary = []
    package_path = os.path.join(project_path, 'package.json')
    if os.path.exists(package_path):
        try:
            with open(package_path, 'r') as f:
                pkg = json.load(f)
            name = pkg.get('name', 'unknown')
            desc = pkg.get('description', '')
            summary.append(f"{name}: {desc}" if desc else f"Project: {name}")
        except:
            pass
    app_path = os.path.join(project_path, 'src/app')
    if os.path.exists(app_path):
        folders = [d for d in os.listdir(app_path) if os.path.isdir(os.path.join(app_path, d))]
        if folders:
            summary.append(f"Features: {', '.join(folders[:3])}")
    summary.append("Angular app with components, services, and routing")
    return ". ".join(s for s in summary if s) + "."

# Scan project
total_size = 0
metadata["project_summary"] = generate_project_summary()
for root, _, files in os.walk(project_path):
    for file in files:
        if 'node_modules' not in root:
            path = os.path.join(root, file).replace(project_path + '/', '')
            file_type = categorize_file(path)
            metadata["files"].append({"path": path, "type": file_type, "summary": ""})
            metadata["stats"]["file_count"] += 1

            full_path = os.path.join(root, file)
            try:
                with open(full_path, 'r', encoding='utf-8') as f:
                    content = f.read()

                file_summary = summarize_content(path, content, file_type)
                metadata["files"][-1]["summary"] = file_summary

                if file_type == 'component':
                    metadata["components"].append(parse_component(path, content, file_summary))
                    metadata["stats"]["component_count"] += 1
                elif file_type == 'service':
                    metadata["services"].append(parse_service(path, content, file_summary))
                    metadata["stats"]["service_count"] += 1
                elif file_type == 'module':
                    metadata["modules"].append(parse_module(path, content, file_summary))
                    if 'routing' in file.lower():
                        metadata["routing"]["module"] = path
                        routes = re.findall(r"path:\s*['\"]([^'\"]+)['\"].*component:\s*(\w+)", content)
                        metadata["routing"]["routes"] = [
                            {"path": r[0], "component": r[1]} for r in routes
                        ]
                elif file_type == 'config' and file == 'package.json':
                    metadata["configs"]["package.json"] = json.loads(content)
                elif file_type == 'asset':
                    metadata["assets"].append({"path": path, "type": file.split('.')[-1], "summary": file_summary})

                total_size += os.path.getsize(full_path) / (1024 * 1024)
            except:
                metadata["files"][-1]["summary"] = summarize_content(path, "", file_type)

metadata["stats"]["size_mb"] = round(total_size, 2)

# Save metadata
os.makedirs(output_folder, exist_ok=True)
metadata_path = os.path.join(output_folder, 'metadata.json')
with open(metadata_path, 'w', encoding='utf-8') as f:
    json.dump(metadata, f, indent=2)
print(f"Created {metadata_path}")

os.makedirs(drive_output_folder, exist_ok=True)
drive_metadata_path = os.path.join(drive_output_folder, 'metadata.json')
with open(drive_metadata_path, 'w', encoding='utf-8') as f:
    json.dump(metadata, f, indent=2)
print(f"Saved to {drive_metadata_path}")

# Preview
print("\nProject Summary:")
print(metadata["project_summary"])
print("\nStats:")
print(f"Files: {metadata['stats']['file_count']}")
print(f"Components: {metadata['stats']['component_count']}")
print(f"Services: {metadata['stats']['service_count']}")
print("\nSample Summaries:")
for f in metadata["files"][:10]:
    print(f"{f['path']}: {f['summary']}")
print("\nNext steps:")
print("1. Check metadata.json in Files tab.")
print("2. Generate migration plan with Claude (next script).")
print("3. Run '!cat nextjs_project/metadata.json' to view.")
````

````python
def migrate_components(state: MigrationState) -> MigrationState:
    spec = state["spec"]
    if not spec.get("components"):
        state["component_output"] = {"files": [], "patterns": {}, "instructions": ""}
        return state

components = spec["components"]
sample = components[:1]  # One representative component
prompt = f"""

You're an expert in Angular and Next.js. Below is a JSON spec for Angular components. Convert one representative component to a Next.js component (components/*.tsx), mapping template to JSX. Output JSON:
Converted file (e.g., components/User.tsx).

Reusable patterns (e.g., *ngFor to .map).

Instructions for others.

Spec:
json

{json.dumps(sample, separators=(',', ':'))}

Instructions:
Map inputs to props, outputs to callbacks.

Convert template bindings (e.g., *ngFor) to JSX.

Use Tailwind for styles.

Keep output ~2-4k tokens.
"""
  response = llm.invoke(prompt)
  try:
  component_output = json.loads(response.content)
  except json.JSONDecodeError:
  component_output = {"files": [], "patterns": {}, "instructions": "Error parsing component output"}
  state["component_output"] = component_output
  state["claude_calls"] += 1
  if len(components) > 10:  # Split if large
  prompt = f"""

You're an expert in Angular and Next.js. Convert remaining Angular components to Next.js components (components/*.tsx) using patterns from previous output. Output JSON:
Converted files.

Updated instructions.

Spec:
json

{json.dumps(components[1:], separators=(',', ':'))}

Previous Patterns:
json

{json.dumps(component_output['patterns'], separators=(',', ':'))}

Instructions:
Apply patterns (e.g., *ngFor to .map).

Keep output ~2-4k tokens.
"""
  response = llm.invoke(prompt)
  try:
      extra_output = json.loads(response.content)
      component_output["files"].extend(extra_output.get("files", []))
      component_output["instructions"] = extra_output.get("instructions", component_output["instructions"])
  except json.JSONDecodeError:
      pass
  state["claude_calls"] += 1
  return state

````

````python
def generate_scaffolding(state: MigrationState) -> MigrationState:
    spec = state["spec"]
    target = {"modules": spec["modules"], "configs": spec["configs"]}
    prompt = f"""
You're an expert in Angular and Next.js. Below is a JSON spec for Angular modules and configs. Generate JSON:
Next.js configs (package.json, next.config.js, tsconfig.json).

Routing files (pages/*.tsx) based on modules.
Keep output ~2k tokens.

Spec:
json

{json.dumps(target, separators=(',', ':'))}

Instructions:
Map module declarations to pages/*.tsx.

Include react, next, typescript in package.json.
"""
  response = llm.invoke(prompt)
  try:
  scaffolding_output = json.loads(response.content)
  except json.JSONDecodeError:
  scaffolding_output = {"files": []}
  state["scaffolding_output"] = scaffolding_output
  state["claude_calls"] += 1
  return state

````

````python
def amplify_patterns(state: MigrationState) -> MigrationState:
    spec = state["spec"]
    service_output = state.get("service_output", {"files": [], "patterns": {}, "instructions": ""})
    component_output = state.get("component_output", {"files": [], "patterns": {}, "instructions": ""})
    scaffolding_output = state.get("scaffolding_output", {"files": []})
    project_dir = state["project_dir"]

os.makedirs(project_dir, exist_ok=True)

# Write scaffolding
for file in scaffolding_output.get("files", []):
    path = os.path.join(project_dir, file["path"])
    os.makedirs(os.path.dirname(path), exist_ok=True)
    with open(path, "w", encoding="utf-8") as f:
        f.write(file.get("content", ""))

# Write converted services
for file in service_output.get("files", []):
    path = os.path.join(project_dir, file["path"])
    os.makedirs(os.path.dirname(path), exist_ok=True)
    with open(path, "w", encoding="utf-8") as f:
        f.write(file.get("content", ""))

# Write converted components
for file in component_output.get("files", []):
    path = os.path.join(project_dir, file["path"])
    os.makedirs(os.path.dirname(path), exist_ok=True)
    with open(path, "w", encoding="utf-8") as f:
        f.write(file.get("content", ""))

# Amplify remaining components
comp_patterns = component_output.get("patterns", {})
for comp in spec.get("components", []):
    if comp["path"] not in [f["path"].replace("components/", "src/app/").replace(".tsx", ".ts") for f in component_output.get("files", [])]:
        target_path = comp["path"].replace("src/app", "components").replace(".ts", ".tsx")
        content = generate_component(comp, comp_patterns)
        path = os.path.join(project_dir, target_path)
        os.makedirs(os.path.dirname(path), exist_ok=True)
        with open(path, "w", encoding="utf-8") as f:
            f.write(content)

# Amplify remaining services
svc_patterns = service_output.get("patterns", {})
for svc in spec.get("services", []):
    if svc["path"] not in [f["path"].replace("pages/api/", "src/app/").replace(".ts", ".service.ts") for f in service_output.get("files", [])]:
        target_path = svc["path"].replace("src/app", "pages/api").replace(".service.ts", ".ts")
        content = generate_service(svc, svc_patterns)
        path = os.path.join(project_dir, target_path)
        os.makedirs(os.path.dirname(path), exist_ok=True)
        with open(path, "w", encoding="utf-8") as f:
            f.write(content)

# Handle CSS
for file in spec.get("other", []):
    if file["path"].endswith(".css"):
        src_path = os.path.join("temp_repo", file["path"])
        target_path = file["path"].replace("src/app", "components").replace(".css", ".module.css")
        if os.path.exists(src_path):
            with open(src_path, "r", encoding="utf-8") as src:
                content = src.read()
            path = os.path.join(project_dir, target_path)
            os.makedirs(os.path.dirname(path), exist_ok=True)
            with open(path, "w", encoding="utf-8") as dst:
                dst.write(content)

return state

````

````python
def generate_component(comp: Dict[str, Any], patterns: Dict[str, str]) -> str:
    template = comp.get("template", {})
    bindings = template.get("bindings", [])
    jsx = ""
    for binding in bindings:
        if "*ngFor" in binding and "*ngFor" in patterns:
            jsx += "items.map(item => <div key={item.id}>{item.name}</div>)"
        elif "ngModel" in binding and "ngModel" in patterns:
            jsx += "<input value={value} onChange={(e) => setValue(e.target.value)} />"
    inputs = ", ".join(comp.get("inputs", []))
    return f"""import React, {{ useState }} from 'react';
export default function {comp.get('class', 'Component')}({{ {inputs} }}) {{
  const [value, setValue] = useState('');
  return <div className="p-4">{jsx}</div>;
}}"""
def generate_service(svc: Dict[str, Any], patterns: Dict[str, str]) -> str:
    endpoints = ""
    for method in svc.get("methods", []):
        if svc.get("http", False) and "http_get" in patterns and method.lower().startswith("get"):
            endpoints += f"""if (req.method === 'GET') {{ const data = await fetch('https://api.example.com/{method.lower()}').then(res => res.json()); res.status(200).json(data); }}
            """
        elif svc.get("http", False) and "http_post" in patterns and method.lower().startswith(("save", "update")):
            endpoints += f"""if (req.method === 'POST') {{ const data = await fetch('https://api.example.com/{method.lower()}', {{ method: 'POST', body: JSON.stringify(req.body) }}).then(res => res.json()); res.status(200).json(data); }}
            """
    return f"""export default async function handler(req, res) {{
  {endpoints}
  res.status(405).json({{ error: 'Method not allowed' }});
}}"""

````

````python
def validate_project(state: MigrationState) -> MigrationState:
    project_dir = state["project_dir"]
    try:
        subprocess.run(["npm", "install"], cwd=project_dir, check=True, capture_output=True)
        subprocess.run(["npm", "run", "build"], cwd=project_dir, check=True, capture_output=True)
        print(f"Next.js project built successfully in {project_dir}!")
    except subprocess.CalledProcessError as e:
        print(f"Validation failed: {e.stderr.decode()}")
    return state

````

````python
workflow = StateGraph(MigrationState)
workflow.add_node("summarize", summarize_codebase)
workflow.add_node("migrate_services", migrate_services)
workflow.add_node("migrate_components", migrate_components)
workflow.add_node("generate_scaffolding", generate_scaffolding)
workflow.add_node("amplify_patterns", amplify_patterns)
workflow.add_node("validate", validate_project)
workflow.set_entry_point("summarize")
workflow.add_edge("summarize", "migrate_services")
workflow.add_edge("migrate_services", "migrate_components")
workflow.add_edge("migrate_components", "generate_scaffolding")
workflow.add_edge("generate_scaffolding", "amplify_patterns")
workflow.add_edge("amplify_patterns", "validate")
workflow.add_edge("validate", END)
app = workflow.compile()

````
